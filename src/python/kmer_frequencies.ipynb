```python
kmer_per_rank = pd.read_csv("F:/Studium/Master/semester5/thesis/data/NCBI/nr90/index_diamond/statistics/kmer_per_rank.tsv", sep="\t", header=0)
#kmer_per_rank = pd.read_csv("F:/Studium/Master/semester5/thesis/data/NCBI/nr_preprocessed_ncbi/index/kmer_per_rank.tsv", sep="\t", header=0)

ranks = ["no rank", "superkingdom", "kingdom", "phylum", "class", "order", "family", "genus", "species"]
kmer_per_rank['rank'] = pd.Categorical(kmer_per_rank['rank'], categories=ranks, ordered=True)
kmer_per_rank.sort_values(by='rank', inplace=True, ascending=False)
kmer_per_rank.reset_index(drop=True)
total_kmers = kmer_per_rank['kmers'].sum()
kmer_per_rank['kmers (%)'] = kmer_per_rank['kmers'] / total_kmers * 100
plt.figure(figsize=(10,6))
plt.bar(kmer_per_rank['rank'], kmer_per_rank['kmers (%)'])
plt.xlabel("Rank")
plt.ylabel("Kmers (%)")
plt.title("Kmer distribution per rank")
plt.show()
print(total_kmers)
kmer_per_rank
```

```python
folder, title = "F:/Studium/Master/semester5/thesis/data/NCBI/nr90/index_diamond_filteredByLikelihood/statistics/", "Kmer values - DIAMOND"
#folder, title = "F:\\Studium\\Master\\semester5\\thesis\\data\\NCBI\\nr90\\index_etchebest\\statistics\\", "Kmer values - Etchbest et al."

kmer_histograms = []
for i, file in enumerate([x for x in os.listdir(folder) if x.endswith("kmer_histogram.tsv")]):
    if i == 0: kmer_histograms.append(pd.read_csv(folder + file, sep="\t", header=0, names=["bucket", file.split("_")[0]]))
    else:  kmer_histograms.append(pd.read_csv(folder + file, sep="\t", header=0, usecols=[1], names=[file.split("_")[0]]))

kmer_histogram = pd.concat(kmer_histograms, axis=1)
kmer_histogram["sum"] = kmer_histogram.iloc[:, 1:].sum(axis=1)
total_kmers = kmer_histogram["sum"].sum()
kmer_histogram["sum (%)"] = kmer_histogram["sum"] / total_kmers * 100
kmer_histogram.loc['column sums'] = kmer_histogram.sum()

column_sums = kmer_histogram.loc['column sums'][1:-2]
column_sums = column_sums.sort_values(ascending=False)
column_sums

plt.figure(figsize=(10,6))
plt.bar(column_sums.index, column_sums)
plt.show()
kmer_histogram
```
